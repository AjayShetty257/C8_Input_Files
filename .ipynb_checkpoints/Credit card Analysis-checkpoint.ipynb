```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
# dim_customers_df = pd.read_csv("dim_customers.csv")
# fact_spends_df = pd.read_csv("fact_spends.csv")

# Merge datasets on 'customer_id'
merged_df = pd.merge(dim_customers_df, fact_spends_df, on='customer_id')

# Demographic classification based on age group
age_group_counts = merged_df['age_group'].value_counts()
sns.barplot(x=age_group_counts.index, y=age_group_counts.values)
plt.title('Distribution of Customers by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.show()

# Demographic classification based on gender
gender_counts = merged_df['gender'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Distribution of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.show()

# Demographic classification based on occupation
occupation_counts = merged_df['occupation'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=occupation_counts.index, y=occupation_counts.values)
plt.title('Distribution of Customers by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()

# Insights based on spending categories
plt.figure(figsize=(10, 6))
sns.countplot(x='category', data=merged_df)
plt.title('Distribution of Spending Categories')
plt.xlabel('Category')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.show()

# Insights based on payment types
plt.figure(figsize=(10, 6))
sns.countplot(x='payment_type', data=merged_df)
plt.title('Distribution of Payment Types')
plt.xlabel('Payment Type')
plt.ylabel('Number of Transactions')
plt.show()
```

# Spending Insights

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Group spending by category
spending_by_category = merged_df.groupby('category')['spend'].sum().sort_values(ascending=False)

# Visualize spending by category
plt.figure(figsize=(10, 6))
sns.barplot(x=spending_by_category.values, y=spending_by_category.index)
plt.title('Total Spending by Category')
plt.xlabel('Total Spending')
plt.ylabel('Category')
plt.show()

# Analyze spending by demographic factors
demographic_factors = ['occupation', 'gender', 'city', 'age_group']
for factor in demographic_factors:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='spend', y='category', data=merged_df, hue=factor, estimator=sum, ci=None)
    plt.title(f'Total Spending by Category and {factor}')
    plt.xlabel('Total Spending')
    plt.ylabel('Category')
    plt.legend(title=factor)
    plt.show()
```

# Key customer segments

```python
import pandas as pd

# Load datasets
dim_customers_df = pd.read_csv("dim_customers.csv")
fact_spends_df = pd.read_csv("fact_spends.csv")

# Merge datasets on 'customer_id'
merged_df = pd.merge(dim_customers_df, fact_spends_df, on='customer_id')

# Calculate total spending and average income for each customer
customer_summary = merged_df.groupby('customer_id').agg(
    total_spending=('spend', 'sum'),
    average_income=('avg_income', 'mean'),
    num_transactions=('spend', 'count')
)

# Calculate spending per transaction
customer_summary['spending_per_transaction'] = customer_summary['total_spending'] / customer_summary['num_transactions']

# Segment customers based on spending behavior and income
high_value_customers = customer_summary[
    (customer_summary['spending_per_transaction'] > customer_summary['spending_per_transaction'].quantile(0.75)) &
    (customer_summary['average_income'] > customer_summary['average_income'].quantile(0.75))
]

# Print high-value customer segments
print("High-Value Customer Segments:")
print(high_value_customers)

# Profile high-value customer segments
high_value_customers_profile = pd.merge(high_value_customers, dim_customers_df, on='customer_id')

# Print profile of high-value customer segments
print("\nProfile of High-Value Customer Segments:")
print(high_value_customers_profile)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize spending distribution of high-value customers
plt.figure(figsize=(10, 6))
sns.histplot(high_value_customers['total_spending'], bins=20, kde=True)
plt.title('Spending Distribution of High-Value Customers')
plt.xlabel('Total Spending')
plt.ylabel('Frequency')
plt.show()

# Visualize income distribution of high-value customers
plt.figure(figsize=(10, 6))
sns.histplot(high_value_customers['average_income'], bins=20, kde=True)
plt.title('Income Distribution of High-Value Customers')
plt.xlabel('Average Income')
plt.ylabel('Frequency')
plt.show()

# Visualize spending per transaction distribution of high-value customers
plt.figure(figsize=(10, 6))
sns.histplot(high_value_customers['spending_per_transaction'], bins=20, kde=True)
plt.title('Spending per Transaction Distribution of High-Value Customers')
plt.xlabel('Spending per Transaction')
plt.ylabel('Frequency')
plt.show()

# Visualize demographic characteristics of high-value customers
plt.figure(figsize=(14, 6))
sns.countplot(data=high_value_customers_profile, x='occupation')
plt.title('Occupation Distribution of High-Value Customers')
plt.xlabel('Occupation')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(data=high_value_customers_profile, x='gender')
plt.title('Gender Distribution of High-Value Customers')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=high_value_customers_profile, x='age_group')
plt.title('Age Group Distribution of High-Value Customers')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(data=high_value_customers_profile, x='city')
plt.title('City Distribution of High-Value Customers')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```
